### **1. Import semua library yang diperlukan: pandas, numpy, scipy.stats.mstats**

```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
```

### **2. Muat dataset: df = pd.read_csv(url_atau_path_lokal)**

```python
df = pd.read_csv('/housing_dirty.csv')
```

### **3. Eksplorasi awal: df.info(), df.describe(), df.isnull().sum()**

```python
df.info()
df.describe()
df.isnull().sum()
```

### **4. Hapus baris duplikat: df.drop_duplicates(inplace=True)**

```python
df.drop_duplicates(inplace=True)
```

### **5. Normalisasi string: .str.strip().str.title() untuk kolom kota, .str.lower() untuk kondisi**

```python
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

### **6. Imputasi missing values: median untuk numerik, modus untuk kategorik**

```python
# Median > Numerik
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())

# Modus > Kategorik
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])
```

### **7. Tangani outlier dengan IQR Fence pada kolom harga_juta dan luas_m2**

```python
for col in ['harga_juta', 'luas_m2']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

### **8. Validasi akhir: df.isnull().sum().sum() harus = 0, df.duplicated().sum() harus = 0**

```python
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)
```

### **9. Ekspor: df.to_csv('housing_clean.csv', index=False)**

```python
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```
