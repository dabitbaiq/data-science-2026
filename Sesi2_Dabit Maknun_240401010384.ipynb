### **2. Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?**

```python
missing = df.isnull().sum()
persentase = (missing / len(df)) * 100

hasil_missing = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persentase
})

print(hasil_missing)
```

Kolom dengan missing value adalah:
1. Kolom age jumlah missing 177 dan persentase 19.865320%
2. Kolom cabin dengan jumlah missing 687 dan persentase 77.104377%
3. Kolom embarked dengan jumlah missing 2 dan persentase 0.224467%

### **3. Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas**

```python
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase penumpang selamat: {persen_selamat:.2f}%")
```

### **4. Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?**

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah = len(wanita_kelas1)
rata_usia = wanita_kelas1['Age'].mean()

print("Jumlah:", jumlah)
print("Rata-rata usia:", rata_usia)
```

Jumlah penumpang wanita adalah 94 dengan rata-rata usia 34.61176470588235

### **5. Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?**

```python
keselamatan = df.groupby('Pclass')['Survived'].mean()
print(keselamatan)
```

Tingkat keselamatan tertinggi yaitu pada (Pclass) 1 dengan tingkat keselamatan 0.629630

### **6. Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik.**

```python
fare_kelas = df.groupby('Pclass')['Fare'].mean()
print(fare_kelas)
```

Menghitung rata-rata tarif tike (Fare) berdasarkan kelas penumpang (Pclass) menunjukkan jika Pclass 1 memiliki rata-rata dengan tarif tertinggi yaitu 84.154687
